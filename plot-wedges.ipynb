```python
import pandas as pd
import numpy as np
import scipy.io

from matplotlib import cm
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.interpolate import griddata
import matplotlib.colors as colors

import seaborn as sns

# import weightedcalcs as wc
```

```python
df = pd.read_csv("labor-supply-results-no-wedge.csv")

dfwedge = pd.read_csv("labor-supply-results-wedge.csv")



```

```python
df.head()
```

```python
fig, ax = plt.subplots(figsize=(10,6), tight_layout = True)

fontsize = 16

ax.plot(100*df['τ'], 100*df['welfare_GHH'], 
        label='Lump Sum', color='darkred', linewidth=5)

ax.plot(100*df['τ'], 100*df['welfare_GHH_labortax'], 
        label='Labor Subsidy', color='darkblue', linewidth=5)

ax.spines["right"].set_visible(False) # This removes the ``spines'', just the right and top
ax.spines["top"].set_visible(False) # ones...

ax.set_xlabel("Tariff Rate", fontsize = fontsize,) # The xlabel
ax.set_ylabel("Welfare Gain (%)", fontsize = fontsize,) # The ylabel

ax.tick_params(axis='both', which='major', labelsize=fontsize)

ax.legend(loc="upper right", fontsize = fontsize, frameon=False)

ax.axvline(4, linewidth = 4, ls = "--", color = 'grey', alpha =0.75)
ax.axhline(0.0, linewidth = 4, ls = "--", color = 'grey', alpha =0.75)

ax.set_title("No Wedges", fontsize = fontsize+2)

ax.set_xlim(-2,60) # This sets the y-limits
ax.set_ylim(-.30,1.25) # This sets the y-limits

plt.savefig("tariff-no-wedge.pdf", bbox_inches = "tight")

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,6), tight_layout = True)

fontsize = 16

ax.plot(100*dfwedge['τ'], 100*dfwedge['welfare_GHH'], 
        label='Lump Sum', color='darkred', linewidth=5)

ax.plot(100*dfwedge['τ'], 100*dfwedge['welfare_GHH_labortax'], 
        label='Labor Subsidy', color='darkblue', linewidth=5)

ax.plot(100*df['τ'], 100*df['welfare_GHH'], 
    color='grey', linewidth=3, alpha = 0.15)

ax.plot(100*df['τ'], 100*df['welfare_GHH_labortax'], 
         color='grey', linewidth=3, alpha = 0.15)

ax.spines["right"].set_visible(False) # This removes the ``spines'', just the right and top
ax.spines["top"].set_visible(False) # ones...

ax.set_xlabel("Tariff Rate", fontsize = fontsize,) # The xlabel
ax.set_ylabel("Welfare Gain (%)", fontsize = fontsize,) # The ylabel

ax.tick_params(axis='both', which='major', labelsize=fontsize)

ax.legend(loc="upper right", fontsize = fontsize, frameon=False)

ax.axvline(4, linewidth = 4, ls = "--", color = 'grey', alpha =0.75)
ax.axhline(0.0, linewidth = 4, ls = "--", color = 'grey', alpha =0.75)

ax.set_title("With Wedges", fontsize = fontsize+2)

ax.set_xlim(-2,60) # This sets the y-limits
ax.set_ylim(-.30,1.25) # This sets the y-limits

# plt.savefig("tariff-wedge.pdf", bbox_inches = "tight")

plt.show()
```
